```python
import hashlib
class Block:
    def __init__(self, data, previous_hash):
        self.data = data
        self.previous_hash = previous_hash
        self.hash = self.calculate_hash()

    def calculate_hash(self):
        block_string = self.data + self.previous_hash
        return hashlib.sha256(block_string.encode()).hexdigest()

#Genesis Block
block1 = Block("Alicia pays Bob $1000" , "0")

#Second Block
block2 = Block("Bob pays Charlie $5" , block1.hash)

#Third Block
block3 = Block("Charlie pays David $2", block2.hash)

print("Block 1")
print("Data", block1.data)
print("Hash", block1.hash)

print("\nBlock 2")
print("Data:", block2.data)
print("Previous Hash:", block2.previous_hash)
print("Hash:" , block2.hash)

print("\nBlock 3")
print("Data:", block3.data)
print("Previous Hash:", block3.previous_hash)
print("Hash:", block3.hash)
```
